```python
#Googled and found how to convert the strings into numbers.
#For example: In [143]: df[['A','B','C']] = df[['A','B','C']].apply(pd.to_numeric, errors='coerce')

df[['Mean Score','Number Scoring Below 65','Percent Scoring Below 65','Number Scoring 65 or Above','Percent Scoring 65 or Above','Number Scoring 80 or Above','Percent Scoring 80 or Above','Number Scoring CR','Percent Scoring CR']] = df[['Mean Score','Number Scoring Below 65','Percent Scoring Below 65','Number Scoring 65 or Above','Percent Scoring 65 or Above','Number Scoring 80 or Above','Percent Scoring 80 or Above','Number Scoring CR','Percent Scoring CR']].apply(pd.to_numeric, errors='coerce')
```

```python
#Trying to parse out the info. from School DBN

df['School DBN'][0][3:]
```

```python
#I needed to create a new column called Borough by splitting the third character from School DBN. Then, I mapped the character to a more human readable name.

df['Borough'] = df['School DBN'].str[2]

borough_mapping = {
    "M": "Manhattan",
    "X": "Bronx",
    "K": "Brooklyn",
    "Q": "Queens",
    "R": "Staten Island"
}
df['Borough'] = df['Borough'].map(borough_mapping)
```

```python
#Verified new column data

df.head(5)
```

```python
#Created a new column called School Number.

df['School Number'] = df['School DBN'].str[3:]
```

```python
#Created a new column called School District.

df['School District'] = df['School DBN'].str[:2]
```

```python
#I wanted to understand the types of levels within 'School Level'

df['School Level'].unique()
```

```python
#I focused only on High School.

df_hs = df[df['School Level'] == 'High school']
```

```python
#What's the higest percentage of high schools that achieve 80 or above. Found out it was 100%

df_hs['Percent Scoring 80 or Above'].max()
```

```python
#which high school actually received 100? Which borough is it in? What about the size of the test takers? It was Queens High School for the Sciense at York College.

df_hs.sort_values(['Percent Scoring 80 or Above'], ascending=[False])
```

```python
# a new data frame -nd-  is created that isolates the data for only high schools. 
nd = df[df['School Level'] == 'High school']
```

```python
#verified

nd.head(5)
```

```python
# I wanted to sort the schools by 'Percent Scoring 80 or Above' in a descending order.

nd.sort_values('Percent Scoring 80 or Above', ascending=False)
```

```python
#I wanted to aggregate by boroughs & year & understand the total number of students scoring 80 or above & the average of percent of those students.
#Plus, I wanted to understand the total number of students tested & the schools.

nd2 = nd.groupby(['Borough','Year'])\
                      .agg(total_students_scoring_80_or_above = ('Number Scoring 80 or Above', 'sum'), total_tested =('Total Tested', 'sum'), avg_percent_scoring_of_80_or_above = ('Percent Scoring 80 or Above','mean'), total_schools = ('School DBN', 'nunique') )
```

```python
nd2
```

```python
#To simplify, I wanted to combine the three years.

nd3 = nd.groupby(['Borough'])\
                      .agg(total_students_scoring_80_or_above = ('Number Scoring 80 or Above', 'sum'), total_tested =('Total Tested', 'sum'), avg_percent_scoring_of_80_or_above = ('Percent Scoring 80 or Above','mean'), total_schools = ('School DBN', 'nunique') )
```

```python
nd4 = nd3.sort_values('avg_percent_scoring_of_80_or_above', ascending=False)
```

```python
#Verified it
nd4
```

```python
#I needed to compare the Queens High School for the Sciense at York College against the borough average.

nd.loc[df['School DBN'] == '28Q687'].groupby(['School Name'])\
                      .agg(total_students_scoring_80_or_above = ('Number Scoring 80 or Above', 'sum'), total_tested =('Total Tested', 'sum'), avg_percent_scoring_of_80_or_above = ('Percent Scoring 80 or Above','mean') )
```

Queens High School for the Sciences at York College has one of the smartest student bodies in NYC public high schools. In fact, 86% of the students score 80 or above from 2015 to 2017, beating all other high schools. As a point of comparison, in Staten Island, the average percentage for those scoring 80 or higher was only 25%, the highest for all five boroughs. Bronx had 13%, the lowest.


As for visualization, I wanted to create a scatter plot that shows the relationship between the total number of students who took the test and the avg. percentage of those who scored 80% and above. The size of the bubble was each school's size of the students who scored 80% and above.

```python
nd5 = nd.groupby(['School Name','Borough'])\
                      .agg(total_students_scoring_80_or_above = ('Number Scoring 80 or Above', 'sum'), total_tested =('Total Tested', 'sum'), avg_percent_scoring_of_80_or_above = ('Percent Scoring 80 or Above','mean') )
```

```python
# Needed to understand what was the index after aggregation
# Used these two functions to get more info: print(list(nd3.index.values)) and print(list(nd3.index.values.tolist()))
# Then reset the index for the new data frame

nd6 = nd5.reset_index()
```

```python
#Verified it

nd6.columns
```

```python
#I googled and found a librar for plotly. So I used "import plotly.express as px" to generate the following chart.

fig = px.scatter(nd6, y='avg_percent_scoring_of_80_or_above', size= 'total_students_scoring_80_or_above', x = 'total_tested', color = 'Borough', hover_name = 'School Name', size_max=60)

fig.show()
      
```
